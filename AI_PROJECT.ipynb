```python
# ────────────────────────────────────────────────────────────────
# CELL 10 — Start Flask API + ngrok tunnel
# Run this AFTER Cell 8 (both pipelines loaded)
# pipeline_rag and pipeline_ft must already exist in memory
# ────────────────────────────────────────────────────────────────

!pip install flask flask-cors pyngrok -q

from flask import Flask, request, jsonify
from flask_cors import CORS
from pyngrok import ngrok, conf
import threading
import time

# ── Set your ngrok auth token ──────────────────────────────────
# Get it free at: dashboard.ngrok.com → Your Authtoken
ngrok.set_auth_token() #paste ur ngrok config authtoken here

# ── Flask app ──────────────────────────────────────────────────
app_server = Flask(__name__)
CORS(app_server)

@app_server.route('/health', methods=['GET'])
def health():
    return jsonify({'status': 'ok'})

@app_server.route('/ask', methods=['POST'])
def ask():
    data  = request.json
    query = data.get('query', '').strip()
    mode  = data.get('mode', 'llama')

    if not query:
        return jsonify({'error': 'empty query'}), 400

    t0 = time.time()
    try:
        # Uses the real pipeline objects loaded in Cell 5 and Cell 8
        if mode == 'llama':
            answer = pipeline_rag.ask(query)
        else:
            answer = pipeline_ft.ask(query)
    except Exception as e:
        return jsonify({'error': str(e)}), 500

    return jsonify({
        'answer':  answer,
        'time_ms': int((time.time() - t0) * 1000)
    })

# ── Start Flask in background thread ──────────────────────────
threading.Thread(
    target=lambda: app_server.run(port=5001, use_reloader=False),
    daemon=True
).start()

# ── Start ngrok tunnel ─────────────────────────────────────────
public_url = ngrok.connect(5001).public_url
print(f"\n✅ Your API is live at: {public_url}")
print(f"\n👉 Copy this URL into index.html:")
print(f'   const API_URL = "{public_url}";')
print(f"\n🔍 Health check (paste in browser to verify):")
print(f"   {public_url}/health")
```
